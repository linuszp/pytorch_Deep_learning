```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

###### data: 数据集
###### forward ：模型
###### loss ： 损失函数 
$$
\text { loss }=(x \cdot w-y)^{2}
$$
###### MSE：模型评价

```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]

def forward(x):
    return x*w
def loss(x,y):
    y_pred = forward(x)
    return (y_pred-y)*(y_pred-y)

w_list = []
mse_list =[]
for w in np.arange(0.0,4.1,0.1):
    print("w=",w)
    l_sum = 0
    for x_val,y_val in zip(x_data,y_data):
        y_pre_val = forward(x_val)
        loss_val = loss(x_val,y_val)
        l_sum += loss_val
        print('\t',x_val,y_val,y_pre_val,loss_val)
    mse_val =l_sum/3    
    print('MSE=',mse_val)
    w_list.append(w)
    mse_list.append(mse_val)    

```

```python
#画图
plt.plot(w_list,mse_list)
plt.xlabel('w')
plt.ylabel("loss")
plt.show()
```
