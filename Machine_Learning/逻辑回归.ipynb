#####  logistic function(sigmod function)
$$
f(x)=\frac{1}{1+e^{-x}}
$$
##### loss function
$$
P(y \mid x)=\hat{y}^{y}(1-\hat{y})^{(1-y)}
$$
![image](https://latex.codecogs.com/gif.latex?%5Cdpi%7B120%7D%20%5Clarge%20loss%20%3D%20%5Cfrac%7B1%7D%7BN%7D%5Csum%20L)

```python
import torch
import torch.nn.functional as F

x_data = torch.Tensor([[1.0],[2.0],[3.0]])
y_data = torch.Tensor([[0],[0],[1]])

class LogisticRegressionModel(torch.nn.Module):
    def __init__(self):
        super(LogisticRegressionModel,self).__init__()
        self.linear = torch.nn.Linear(1,1)
    def forward(self,x):
        y_pred = F.sigmoid(self.linear(x))
        return y_pred
model = LogisticRegressionModel()       
    #BSE
criterion = torch.nn.BCELoss(size_average=False) #二分类交叉熵
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01) # model.parameters()自动完成参数的初始化操作
for epoch in range(10000):
    y_pred = model(x_data)
    loss = criterion(y_pred,y_data)
    print(epoch,loss.item())

    optimizer.zero_grad()#权重梯度归0，开始反向传播
    loss.backward()
    optimizer.step()

print('w = ', model.linear.weight.item())
print('b = ', model.linear.bias.item())
x_test = torch.Tensor([[4.0]])
y_test = model(x_test)
print('y_pred = ', y_test.data)   
```
